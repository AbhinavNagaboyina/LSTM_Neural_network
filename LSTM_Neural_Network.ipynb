```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
vocab_size=7000
```

```python
msgs= X.copy()
```

```python
msgs.reset_index(inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
msgs['title'][0]
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
corpus=[]

for i in range (0, len(msgs)):
  review= re.sub('[^a-zA-Z]',' ',msgs['title'][i])
  review= review.lower()
  review= review.split()
  review= [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review= ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
onehot_repr= [one_hot(words,vocab_size) for words in corpus]
```

```python
onehot_repr
```

```python
sen_len= 25
embeds= pad_sequences(onehot_repr, padding='pre', maxlen= sen_len)
embeds
```

```python
len(embeds)
```

```python
embed_vec_features=50
model= Sequential()
model.add(Embedding(vocab_size,embed_vec_features,input_length=sen_len))
model.add(LSTM(100))
model.add(Dense(1, activation= 'sigmoid'))
model.compile(loss='binary_crossentropy', optimizer= 'adam', metrics=['accuracy'])
model.summary()
```

```python
len(embeds)
```

```python
y.shape
```

```python
import numpy as np
X_final= np.array(embeds)
y_final= np.array(y)
```

```python
X_final.shape
```

```python
y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X_final, y_final, test_size= 0.25, random_state= 42)
```

```python
model.fit(X_train, y_train, validation_data= (X_test, y_test), epochs= 10, batch_size= 64)
```

```python
y_pred = (model.predict(X_test) > 0.5).astype("int32")
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from tensorflow.keras.layers import Dropout
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(vocab_size,embed_vec_features,input_length=sen_len))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, validation_data= (X_test, y_test), epochs= 10, batch_size= 64)
```

```python
y_pred = (model.predict(X_test) > 0.5).astype("int32")
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
